## 課題１：振幅スペクトルの描画

`sound/ex-Violin.wav`を開いて、その波形の振幅スペクトルを描画し、このノートブックが置かれているディレクトリ内に`ex2_1.png`という名前の画像として保存してください。   
振幅値$x$は$20 \log_{10}(x)$によりデシベル値に変換してください。  
サンプリングレートはwaveオブジェクトから`wave.getframerate()`を使って受け取ってください。   
x軸は周波数（単位はHz)、y軸は振幅のdB値とします。   
軸メモリと軸ラベルを付けてください。   
次のようなグラフになったらOKです。
<img src="http://www.hal.t.u-tokyo.ac.jp/~yamakata/lecture/mediaproc/mediaproc2/fig/ex2_ans1.png" width="640">


```python
%matplotlib inline
import wave
import numpy as np
import matplotlib.pyplot as plt

wavfile = wave.open('sound/ex-Violin.wav', 'rb')
#Read and return at most nframes of audio as a byte object.
x = wavfile.readframes(wavfile.getnframes()) 
#Return sample rate
sampling_rate = wavfile.getframerate()
#Return sample width (in bytes)
sample_size = wavfile.getsampwidth() 
wavfile.close()

#print(sample_size)

#Convert binary data to int type
x = np.frombuffer(x, dtype= "int16")
#Normalization
x = x/(2**(8 * sample_size) / 2)

### Fast Fourier Transform with FFT
X = np.fft.fft(x)  
# The vector after Fourier transform has the same number of elements as the waveform
#print("len(x): ", len(x), " len(X): ", len(X))

#Amplitude transformation
amplitude = 20*np.log10(np.abs(X))
x_axis=[i/len(X)*sampling_rate for i in range(0,int(len(X)/2)-1)]

#Specify size
fig = plt.figure(figsize=(12, 5)) 
ax = plt.subplot() 
ax.plot(x_axis[0:int(len(X)/2-1)],amplitude[0:int(len(X)/2-1)])
ax.set_xlabel("fraquency [Hz]")
ax.set_ylabel("amplitude [dB]")
ax.grid()
#Save image
plt.savefig('fig/Sex2_1.png')
```

## 課題２：スペクトログラムの描画

ヴァイオリンの音楽`sound/Violin-music.wav`のスペクトログラムを描画し、このノートブックが置かれているディレクトリ内に`ex2_2.png`という名前の画像として保存してください。   
スペクトログラム計算時のパラメータは以下の通りです（これは教材と同じです）。
- 窓幅は1024
- 移動窓における重なり（オーバーラップ）は256フレーム
- 窓関数はハミング窓

次のようなグラフになったらOKです。
<img src="http://www.hal.t.u-tokyo.ac.jp/~yamakata/lecture/mediaproc/mediaproc2/fig/ex2_ans2_2020A2.png" width="640">

```python
%matplotlib inline

import wave
from pylab import *
import numpy as np

wavfile = wave.open("sound/ex-Violin-music.wav" , "rb" )

x = wavfile.readframes(wavfile.getnframes())
x = np.frombuffer(x, dtype= "int16") 
sampling_rate = wavfile.getframerate()
#print(sampling_rate)
wavfile.close()

N = 1024
# Humming window used in FFT
hammingWindow = np.hamming(N)

#Draw spectrogram
specgram(x, NFFT=N, Fs=sampling_rate, noverlap=256, window=hammingWindow,cmap=cm.jet) 
# Sampling rate  # Specifying a window function
xlabel("time [second]")
ylabel("frequency [Hz]")
plt.savefig('fig/ex2_2.png') # Save the figure as an image
```
